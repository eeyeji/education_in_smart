```python
import numpy as np
import pandas as pd
import random
```

```python
data = np.loadtxt("ratings.txt", delimiter='::', dtype = 'int')

print("======================================================")
print(data)
print("======================================================")

#크기 확인
print(data.shape)
print("======================================================")

#차원 확인
print(data.ndim)
print("======================================================")

#원소 개수 확인
print(data.size)
print("======================================================")

#1번 아이디를 가진 사람의 정보만 출력
#               행 열
print(data[data[:, 0]==1])
print("======================================================")

#id 정보만 가져오기
id = data[:, 0]
print(id)
print("======================================================")

#전체 사용자 평점 평균 구하기
rating = data[:, 2]
print(rating)
print(np.mean(rating))
print("======================================================")
```

##### 1번 사용자의 평점 평균 구하기

```python
#1. ID정보 가져오기
print("======================================================")
id_1 =data[data[:, 0]==1]
print(id_1)
print("======================================================")

#2. 인덱싱한 자료에서 평점만 가져오기
rating_id1 = id_1[:, 2]
print(rating_id1)
print("======================================================")

#3. 가져온 평점 평균 구하기
print(rating_id1.mean())
print("======================================================")
```
