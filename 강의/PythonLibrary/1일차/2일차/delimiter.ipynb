##### BMI구하기
- delimiter: 구분기호

```python
#구분기호 사용해서 구분해준다음 텍스트 파일 불러오기
#변수명 = np.loadtxt("불러오고 싶은 파일명", delimiter = "구분기호") 
data = np.loadtxt("height_weight.txt", delimiter = ",")
print("==================================================================")
print(data)
print("==================================================================")
print(data.shape)
print("==================================================================")

#데이터를 키와 몸무게로 나누기
#키
height_cm = data[0, :]
print(height_cm)
print("==================================================================")

#몸무게
weight_kg = data[1, :]
print(weight_kg)
print("==================================================================")

#키 단위 바꾸기: cm -> m
height_m = height_cm / 100
print(height_m)
print("==================================================================")

#BMI구하기
BMI = weight_kg / (height_m * height_m)
print(BMI)
print("==================================================================")
```
