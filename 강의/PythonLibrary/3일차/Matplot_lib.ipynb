##### 여러 개의 선 그어주고 각각의 선한테 이름 붙여주기

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

plt.plot(x, label='x')
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend()
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
z=[3,6,9,12]

#plot(선) 이름 지정
plt.plot(x, label='가')
plt.plot(y, label='나')
plt.plot(z, label='다')
plt.legend()

#축 이름 지정
plt.xlabel('x축')

#y축 이름 지정
#                 글씨 회전 시켜주기(0~360도)
plt.ylabel('y축').set_rotation(0)

#차트 이름 지정
plt.title('범례표시')
plt.show()
```
