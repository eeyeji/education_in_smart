```python
test=pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc-kr')
test.head()
```

- Traffic_Accident_2017.csv 데이터를 이용하여 주간과 야간에 발생한 사건에 대해 주어진 조건을 지켜 다음과 같이 시각화 하시오
    - 조건1. 위도는 LAT으로, 경도는 LON으로 표기할 것
    - 조건2. X축의 제한은 123부터 133까지로 줄 것
    - 조건3. 투명도는 0.1로 줄 것
    - 조건4. 주간의 데이터와 야간의 데이터가 구분될 수 있게 할 것.

```python
test=test.rename(columns = {'경도' : 'LON', '위도' : 'LAT'})
test=test.replace({'주야':{'주간':'Day','야간':'Night'}})
sns.scatterplot(data=test, x='LON', y='LAT', hue='주야', alpha=0.1)
plt.xlim(123, 133)
plt.show()
```
