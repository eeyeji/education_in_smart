- 목표: 영화상영시간이 200분 이상인 영화만 필터링 하기

```python
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
booleans
```

```python
#데이터프레임 만들기 전에 리스트를 Series로 변환
is_long=pd.Series(booleans)
is_long
```

```python
movies[is_long]
```

```python
#위 과정을 더 간단하게 코드짜기
is_long2 = movies.duration>=200
movies[is_long2]
```

```python
movies[movies.duration>=200]
```
