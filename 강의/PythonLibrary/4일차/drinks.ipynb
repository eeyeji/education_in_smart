```python
drinks100=drinks[(drinks.beer_servings>=100)&(drinks.spirit_servings>=100)&(drinks.wine_servings>=100)]
drinks100
```

```python
drinks100.set_index('country', inplace=True)
drinks100
```

```python
avg=drinks100.mean()
avg
```

```python
avg[:3].plot(kind='pie', ylabel='상위 100 평균', autopct='%.2f', explode=[0.10, 0, 0])
```

```python
#전체 데이터의 맥주 평균 소비량
drinks.beer_servings.mean()
```

```python
#아프리카 대륙의 평균 소비량
drinks[drinks.continent=='Africa'].wine_servings.mean()
```

```python
#각 대륙별 맥주 평균 소비량
drinks.groupby('continent').beer_servings.mean()
```

```python
#각 대륙별 맥주 최고 소비량
drinks.groupby('continent').beer_servings.max()
```

```python
#다양한 통계치를 한꺼번에 표현하고 싶을 때
#                                        여러개볼꺼야
data=drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
data.plot()
```
