```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from matplotlib import font_manager, rc  
font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

##### GDP 실습

```python
years=[1950,1960,1970,1980,1990,2000,2010]
gdp=[67.0, 80.0, 257.0, 1686.0, 6505, 11685.3, 22105.3]
```

- 선 그래프 만들기

```python
plt.plot(years, gdp, ls = '-', marker ='o', lw=5, c='r', ms=6, mec='k', mfc = 'k', mew=3)

plt.title("GDP per capita")
plt.xlabel('년도')
plt.ylabel('dollars').set_rotation(0)
plt.savefig("gdp_per_capita.png", dpi=600)

plt.show()
```

- y = 2x 그래프 만들기

```python
x=[x for x in range(-10, 10)]
y=[2*t for t in x]

plt.plot(x, y, ls = '-', marker ='o', lw=2, c='r', ms=4, mec='k', mfc = 'k', mew=3)
plt.axis([-20, 20, -20, 20])

plt.title("그래프")
plt.xlabel('x축')
plt.ylabel('y축').set_rotation(0)

plt.show()
```

- 1차, 2차, 3차 함수

```python
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]

plt.plot(x, x, label='1번째 함수')

plt.plot(x, y, label='2번째 함수')

plt.plot(x, z, label='3번째 함수')

plt.xlabel('x label')
plt.ylabel('y label')


plt.title("여러 함수 모음")
plt.legend()
plt.show()
```
