```python
#유튜브: https://www.youtube.com/user/koreanenglishman
!pip install webdriver_manager

from tqdm import tqdm_notebook as tqdm

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

from bs4 import BeautifulSoup as bs

#스크롤링
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome(ChromeDriverManager().install())

#유튜브: 티타임즈tv
driver.get("https://www.youtube.com/c/%ED%8B%B0%ED%83%80%EC%9E%84%EC%A6%88TV/videos")
driver.maximize_window()



#------------------------------------------------------------------------------------------
#스크롤링
body = driver.find_element_by_css_selector('body')

#body.send_keys(Keys.END)

name_list = []
view_list = []

while True:

    #현재화면? 객체화
    last = bs(driver.page_source, 'html.parser')
    
    #제목: #video-title -> css_selector가 공유되서 사용되고 있음
    name = driver.find_elements_by_css_selector('#video-title')
    for i in name:
        print(i.text)
        name_list.append(i.text)
 
    #조회수: #metadata-line > span:nth-child(1)
    view = driver.find_elements_by_css_selector('#metadata-line > span:nth-child(1)')
    for j in view:
        
        #조회수: String -> Int
        #만, 천, ., 회 지워주고 변환시켜주기
        String = j.text
        String = String.replace('조회수', '')
        String = String.replace('천', '000')
        String = String.replace('만', '0000')
        String = String.replace('.', '')
        String = String.replace('회', '')
        print(String)
        String=int(String)
        view_list.append(String)



    time.sleep(1)

    for m in range(10):
        body.send_keys(Keys.END)
    #스크롤 10번 한 후의 화면?을 객체화
    new = bs(driver.page_source, 'html.parser')

    #객체화 한 현재화면?과 스크롤한 화면?이 같을 경우 멈춤
    if last == new:
        break
    else:
        pass

print("끝")

data= {'제목' : name_list, '조회수' : view_list}
youtube = pd.DataFrame(data)

#정렬해서 데이터프레임화: 내림차순
#주의: 재정의 필수!
youtube = youtube.sort_values(by=['조회수'], ascending = False)

#?
youtube = youtube.reset_index(drop=True)

youtube

```
