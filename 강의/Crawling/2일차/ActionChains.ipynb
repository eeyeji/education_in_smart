```python
!pip install webdriver_manager

from selenium import webdriver
from webdriver_manager .chrome import ChromeDriverManager
import pandas as pd
import time

#한솥도시락 홈페이지
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.hsd.co.kr/")

#최대화
driver.maximize_window()

#메뉴 xpath주소
menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
#menu.click()

#액션체인 선언
from selenium.webdriver.common.action_chains import ActionChains
actions = ActionChains(driver)

#마우스 커서 옮기기
#1. 원래 위치: 메뉴
actions.move_to_element(menu)

#2.전체메뉴 클릭
total_menu = driver.find_element_by_xpath('//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')
actions.click(total_menu)


#액션체인이 끝. 실행
actions.perform()

#더보기버튼-while문
while True:
    try:
        btn = driver.find_element_by_css_selector('#btn_more > span > a')
        btn.click()
        time.sleep(1)

    except:
        break

#메뉴이름 div.item-text>h4
#주의: 여러개 뽑을 때는 elements!!!!!
# 아래도 가능 클래스가 하나뿐이라?
# menu_name = driver.find_elements_by_css_selector('h4')
menu_name = driver.find_elements_by_css_selector('h4.h.fz_03')
for i in menu_name:
    print(i.text)


# 메뉴가격
price = driver.find_elements_by_css_selector('div.item-price>strong')
for i in price:
    print(i.text)


menu_name_list = []
for i in menu_name:
    menu_name_list.append(i.text)  

price_list = []
for i in price:
    price_list.append(i.text)

data = {'메뉴명' : menu_name_list, '가격' : price_list}

food = pd.DataFrame(data)
food.set_index('메뉴명')
```
