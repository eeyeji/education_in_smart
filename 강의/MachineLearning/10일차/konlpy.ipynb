```python
from collections import Counter
import pandas as pd
import numpy as np
```

- 단어 빈도 계산

```python
df = pd.read_csv('ratings.txt', delimiter='\t')
df
```

```python
x = df['document'][:1000]
```

```python
n=[]
for i in x:
    n.append(i)
```

```python
okt = Okt()
noun = okt.nouns(str(n))
count = Counter(noun)
```

```python
noun_list = count.most_common(100)
for i in noun_list:
    print(i)
```

- 단어리스트 만들기

```python
#word_list = []
word_list = df['document'][:3000]
```

```python
tag=[]
for s in word_list:
    #형태소 분석한 값 저장
    morph = okt.pos(s)
    #
    tag.append(morph)
```

```python
#명사추출
noun_list = []
for s in tag:
    for word, tag in s:
        if tag in ['Noun']:
            noun_list.append(word)
```

```python
noun_list
```

- 두글자 이상 단어만 추출

```python
noun_list = [n for n in noun_list if len(n)>1]
noun_list[:10] 
```

- 단어별 개수 세기

```python
counts = Counter(noun_list)
tags = counts.most_common(1000)
tags
```

##### 워드클라우드 생성

```python
#워드클라우드용 마스크를 생성하기 위한 라이브러리
from PIL import Image

#워드클라우드 import
from wordcloud import WordCloud

#데이터를 시각화 해주는 라이브러리
import matplotlib.pyplot as plt

image_file = 'alice.png'
img_file=Image.open(image_file)
alice_mask = np.array(img_file)

plt.figure(figsize=(8,8))
plt.imshow(alice_mask)
plt.axis('off')
```

```python
wordcloud = WordCloud(font_path= '나눔손글씨 버드나무.ttf',
                        background_color='white',
                        max_words=1000,
                        mask=alice_mask)
wc=wordcloud.generate(str(tags))
plt.figure(figsize=(12,12))
plt.imshow(wc)
plt.axis('off')
```
