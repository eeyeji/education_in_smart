```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import time
from collections import Counter
import pickle
from tqdm import tqdm

import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc  
font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
#(빈)크롬창 열기
driver = webdriver.Chrome(ChromeDriverManager().install())

#url_list만들기
url_list = []

#sbs뉴스: 최신기사 페이지 별 창 열기
for i in range(1, 2):
    page_url = f'https://news.sbs.co.kr/news/newsflash.do?plink=NEW&cooper=SBSNEWSMAIN&pageIdx={i}'
    driver.get(page_url)
    driver.maximize_window()
    time.sleep(1)
    
    #기사 리스트(행) 별 url 저장-href 
    soup = bs(driver.page_source, 'html.parser')
    url = soup.select('a.news')
    for j in url:
        url_list.append('https://news.sbs.co.kr/'+j['href'])
```

```python
title_list=[]
contents_list=[]

for i in url_list:
    driver.get(i)
    time.sleep(1)
    soup = bs(driver.page_source, 'html.parser')
    #제목
    title = soup.select('h2.blind.ir')
#    for i in title:
#         title_list.append(i.text)
    title_list.append(title[0].text)
    
    #내용
    contents = soup.select('div.text_area')
#     for i in content:
#         contents_list.append(i.text)

#리스트에 쌓인 내용만 뽑기위해서 인덱스 사용, 즉 []표시 지우기 위해! 
    contents_list.append(contents[0].text)
```
